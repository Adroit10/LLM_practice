## Embedding Techniques

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is an example of OpenAI embedding"
query =  embeddings.embed_query(text) # dimensions = 1024
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader = TextLoader("speech.txt")
docs = loader.load()
splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_docs = splitter.split_documents(docs)

from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(documents=final_docs,embedding=embeddings)
query = ""
retrieved_results = db.similarity_search(query)
```

## HuggingFace Embeddings

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["HF_TOKEN"] = os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text = "This is a test document"
query_res = embeddings.embed_query(text)
query_res
```

```python
len(query_res)
```

```python
doc_res = embeddings.embed_documents([text,"This is not a test document."])
doc_res
```

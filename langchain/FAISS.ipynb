FAISS (Facebook Ai simmilarity search) for similarity seaarch and clustering of dense vectora

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader= TextLoader("speech.txt")
docs = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000,chunk_overlap = 30)
docs = text_splitter.split_documents(docs)
docs
```

```python
embeddings = OllamaEmbeddings(model="nomic-embed-text")
db = FAISS.from_documents(docs,embeddings)
db
```

```python
query = "How does thespeaker describe the desired outcome of the war?"
docs = db.similarity_search(query)
docs
```

```python
docs[0].page_content
```

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
docs[0].page_content
```

```python
docs_score = db.similarity_search_with_score(query)
docs_score
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
db.similarity_search_by_vector(embedding_vector)
```

```python
## saving and loading
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index",embeddings,allow_dangerous_deserialization=True)
docs = new_db.similarity_search(query)
```
